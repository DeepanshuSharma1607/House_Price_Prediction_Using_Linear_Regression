```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dv=pd.read_csv('Housing.csv')
print(dv.head())
print(dv.size)
```

```python
x=dv['area']
y=dv['price']
print(x.size,y.size)
plt.scatter(x,y,marker='x')
plt.show()
```

```python
x_train=np.array(x)
y_train=np.array(y)
m=x_train.shape[0] # return the lenghth of array
m=x_train.shape    # return the tupple with entry of each dimension
print(m)
```

```python
m=len(x_train)
m
```

```python
x_i=x_train[:]
y_i=y_train[:]
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
x_min,x_max=np.min(x_train),np.max(x_train)
y_min,y_max=np.min(y_train),np.max(y_train)
w_est=((y_max-y_min)/(x_max-x_min))
w_range=w_est*5
print(w_range)
w_i=np.linspace(-w_range,w_range,100)
w_i.shape
```

```python
b_est = y_min - w_est * x_min

b_range = abs(y_max - y_min)

b_i = np.linspace(b_est - b_range, b_est + b_range, 100)
```

```python
J=np.zeros((len(w_i),len(b_i)))
for i,w in enumerate(w_i):
    for j,b in enumerate(b_i):
        prediction=w*x+b
        J[i,j]=(1/(2*m))*np.sum((prediction-y)**2)
J_min_index=np.unravel_index(np.argmin(J),J.shape)
plt.plot(w_i,J,'r')
plt.show()
```
