Using scikit learn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('2018.csv')
data=data.dropna(subset=['GDP per capita','Score'])
print(data.columns)
```

```python
x=data[['GDP per capita']].values
y=data[['Score']].values
```

```python
plt.scatter(x,y,color='blue')
plt.grid(True)
plt.xlabel('GDP per capita')
plt.ylabel('Score')
plt.show()
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
x_new=[[1.305]]
print(model.predict(x_new))
```

```python
w=model.coef_      # w
b=model.intercept_# b
y_pred=model.predict(x)
plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='green')
plt.xlabel('GDP per capita')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```
